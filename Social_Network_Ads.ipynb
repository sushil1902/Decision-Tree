# Hyperparameter Tuning in Decision Tree using GridSearchCV

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/sushil1902/Decision-Tree/main/Social_Network_Ads.csv')
data.head()
```

```python
data['Gender'].replace({'Male':0,'Female':1},inplace = True)
```

```python
data.head()
```

```python
x = data.iloc[:,1:4].values
y = data.iloc[:,-1].values
```

```python
x
```

```python
y
```

# Scaling of Independent Variable

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

# Apply fit and Transform in X

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Implementation of Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
```

```python
clf1.fit(x_train,y_train)
```

```python
y_pred_train = clf1.predict(x_train)
y_pred_test = clf1.predict(x_test)
```

# Evaluation of Model

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred_train),'|' ,accuracy_score(y_test,y_pred_test))
```

There is overfiting

```python
from sklearn.tree import DecisionTreeClassifier
clf2 = DecisionTreeClassifier(criterion='entropy',max_depth=4)
# Use hyperparameter entropy & max_depth to 4 instead of 3
```

```python
clf2.fit(x_train,y_train)
```

```python
y_pred_train = clf2.predict(x_train)
y_pred_test = clf2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred_train),'|' ,accuracy_score(y_test,y_pred_test))
```

 * Overfitting is reduced in clf2 to some extent . However it's still overfitting.

```python
param_dist ={'criterion' : ['gini','entropy'],'max_depth' : [1,2,3,4,5,6,7,None]}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(clf1,param_grid=param_dist,cv=10, n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
grid.best_params_
```

# Another example

```python
from sklearn import svm,datasets
from sklearn.model_selection import GridSearchCV
iris = datasets.load_iris()
parameters = {'kernel':('linear','rbf'),"C":[1,10]}
svc = svm.SVC()
clf = GridSearchCV(svc,parameters)
clf.fit(iris.data,iris.target)
gr = GridSearchCV(estimator=svc,param_grid={'C':[1,10],'kernel':('linear','rbf')})
gr.fit(iris.data,iris.target)
print(gr.best_params_)
print(gr.best_score_)
```

# Randomized SearchCV

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
iris = load_iris()
logistic = LogisticRegression(solver='saga', tol=1e-2, max_iter=200,
                              random_state=0)
distributions = dict(C=uniform(loc=0, scale=4),
                     penalty=['l2', 'l1'])
clf = RandomizedSearchCV(logistic, distributions, random_state=0)
search = clf.fit(iris.data, iris.target)
search.best_params_
```
